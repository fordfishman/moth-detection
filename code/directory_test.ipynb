```python
image_paths = []

for night_folder in os.listdir():
    
    # macs have .DS_Store files we need to ignore
    if not night_folder.startswith('.'):
    
        path = f'{os.getcwd()}/{night_folder}'
        
        image_files = os.listdir(path)
        
        for image_name in image_files:
            
            if image_name.lower().endswith('.jpg') or image_name.lower().endswith('.jpeg'):
                
                image_paths.append(f'{path}/{image_name}')

image_paths[:5] # show first 5 paths
```
