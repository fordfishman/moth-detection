```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data = df_combined1[['box_SA', 'cont_SA', 'B_cont_mean', 'G_cont_mean', 'R_cont_mean', 'H_cont_mean',
       'S_cont_mean', 'V_cont_mean']].copy()
row_ids = list(df_combined1['id'])
row_group = list(all_data.group_name)
data.head()
```

```python
def standardize(x):

  return (x - x.mean(axis=0))/x.std(axis=0)

```

```python
data_std = standardize(data)

# weighting
w = np.array([0.5, 0.5, 1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
data_w = data_std*w
```

```python
data_w.shape
```

```python
len(row_ids)
```

```python
from scipy.cluster.hierarchy import fcluster

clusters = linkage(data_w, method='ward', metric='euclidean')
fcluster(clusters, 0.8, criterion='distance')

# compare to clusters that forrest makes
# check to see if neighbors are in the same group
# as what forrest makes and count?
# quantify error - figure it out
# probably some ML multiclass metric - maybe not

# take forrest clustering as ground truth
# make fake distances between moths - 0 if
# same cluster, large number if not, compare to distance
# in linkage map, smaller difference in distance is good
```

```python


clusters
```

```python
plt.figure(figsize=(13, 12))
dendrogram(
    clusters,
    orientation='right',
    labels=row_group,
    distance_sort='descending',
    show_leaf_counts=False,
    leaf_font_size=10,
    color_threshold = 0.8
)
plt.title('Standardizing after combined, avg and dom color')
plt.show()
```

# PCA and Linear Discriminant Analysis

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

pca_data = all_data.copy().sort_values(by='broad_group')
# all data
x_data = pca_data.drop(columns=['id','h_63.5', 'h_84.7', 'h_105.9', 'h_127.1', 'h_148.2', 'h_169.4', 'h_190.6','image_id', 'group_name', 'broad_group'])
# avg and dominant data
# x_data = pca_data[['box_SA', 'cont_SA', 'B_cont_mean', 'G_cont_mean', 'R_cont_mean', 'H_cont_mean', 'S_cont_mean', 'V_cont_mean']]
# hue band data
# x_data = pca_data[['box_SA', 'cont_SA', 'h_0.0',	'h_21.2',	'h_42.4',	'h_211.8', 'h_232.9', 'h_254.1',	'h_275.3', 'h_296.5', 'h_317.6', 'h_338.8']]

X = standardize(x_data)


y_data = pca_data['broad_group'].copy()



y_data[~((y_data < 3) | (y_data==13)| (y_data==15))] = -99

# y = pca_data['broad_group']
y = y_data

# target_names = pca_data['broad_group'].unique()
target_names = y_data.unique()

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Percentage of variance explained for each components
print(
    f"explained variance ratio (first two components): {pca.explained_variance_ratio_}"
)

colors = {
    0: {'id': 'stain',        'color':'blue'},
    1: {'id': 'macaria',      'color':'green'},
    2: {'id': 'black beetle', 'color':'black'},
    13:{'id': 'spongy moth',  'color':'yellow'},
    15:{'id': 'rosy maple',   'color':'brown'},
    -99:{'id':'other','color':'grey'}
}

default = {'id':'','color':'grey'}

plt.figure()

lw = 2

for i in target_names:
    label = colors.get(i,default)['id']
    color = colors.get(i,default)['color']
    plt.scatter(
        X_r[y == i, 0], X_r[y == i, 1],alpha=0.8, color=color, lw=lw, label=label
    )
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left", shadow=False, scatterpoints=1)
plt.title("PCA of moth dataset")

plt.figure()
for i in target_names:
    label = colors.get(i,default)['id']
    color = colors.get(i,default)['color']
    plt.scatter(
        X_r2[y == i, 0], X_r2[y == i, 1], alpha=0.8, color=color, label=label
    )
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left", shadow=False, scatterpoints=1)
plt.title("LDA of moth dataset")

plt.show()
```
